## 1\. Preço do diamante por tipo de corte

```
import seaborn as sns
data = sns.load_dataset("diamonds")
filtro_data = data[['cut', 'price']]
filtro_data.head()
```

```
# gráfico do exercício 1
from matplotlib import pyplot as plt

plt.figure(figsize=(10, 6))
filtro_data.groupby('cut')['price'].mean().plot(kind='bar')

plt.title('Preço do diamante por tipo de corte')
plt.xlabel('Tipo de corte')
plt.ylabel('Preço médio')

plt.show()


```

## 2\. Número de passageiros em dezembro por ano

```
import seaborn as sns

data = sns.load_dataset("flights")

data.head()
```

```
# gráfico do exercício 2

flights = data
flights_dec = flights[flights['month'] == 'Dec']
with sns.axes_style('whitegrid'):
  grafico = sns.lineplot(x='year', y='passengers', data=flights_dec)
  plt.title('Número de passageiros em dezembro por ano')
  plt.xlabel('Ano')
  plt.ylabel('Número de passageiros')
```

## 3\. Numero de passageiros por mês entre 1949 e 1959

```
import seaborn as sns

data = sns.load_dataset("flights")
data.head()
```

```
# gráfico do exercício 3

flights = data.query('year >= 1949 and year <= 1959')
with sns.axes_style('whitegrid'):
  grafico = sns.lineplot(x='month', y='passengers', hue='year', data=flights)
  plt.title('Número de passageiros por mês entre 1949 e 1959')
  plt.xlabel('Mês')
  plt.ylabel('Número de passageiros')
```
