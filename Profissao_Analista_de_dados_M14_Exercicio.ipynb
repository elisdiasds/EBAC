## 1\. Preço do diamante por tipo de corte

```
import seaborn as sns
data = sns.load_dataset("diamonds")
data
```

```
import numpy as np
from matplotlib import pyplot as plt

plt.figure(figsize=(10, 6))
sns.barplot(x="cut", y="price", data=data, palette="magma", estimator=np.mean)

plt.title('Preço do diamante por tipo de corte')
plt.xlabel('Tipo de corte')
plt.ylabel('Preço médio')

plt.show()

```

**Insight do gráfico 1**: O Gráfico representa o preço médio de diamantes de acordo com o tipo de corte.
Os cortes **Premium** e **Fair** tendem a ter os preços mais elevados.
**Ideal** tem um mais baixo e os cortes **Very Good** e **Good** tem preços intermediários.

## 2\. Número de passageiros em dezembro por ano

```
import seaborn as sns

data = sns.load_dataset("flights")

data.head()
```

```
# gráfico do exercício 2

flights = data
flights_dec = flights[flights['month'] == 'Dec']
with sns.axes_style('whitegrid'):
  grafico = sns.lineplot(x='year', y='passengers', data=flights_dec)
  plt.title('Número de passageiros em dezembro por ano')
  plt.xlabel('Ano')
  plt.ylabel('Número de passageiros')
```

**Insight do gráfico 2**:
No gráfico acima temos o seguinte cenário o aumento considerável do número de passageiros com o passar dos anos num intervalo de 10 anos.
Tivemos um aumento significativo no número de passageiros entre 1955 e 1957 com um queda em 1958, tendo o maior pico em 1960.

## 3\. Numero de passageiros por mês entre 1949 e 1959

```
import seaborn as sns

data = sns.load_dataset("flights")
data.head()
```

```
# gráfico do exercício 3

flights = data.query('year >= 1949 and year <= 1959')
with sns.axes_style('whitegrid'):
  grafico = sns.lineplot(x='month', y='passengers', hue='year', data=flights)
  plt.title('Número de passageiros por mês entre 1949 e 1959')
  plt.xlabel('Mês')
  plt.ylabel('Número de passageiros')
```

**Insight do gráfico 3**:
O Gráfico mostra uma distribuição de número de passageiros por mês entre 1949 e 1959.

Com um aumento gradativo com o passar dos anos, tendo um número menor em 1950 e 1952 e um pico nos últimos anos de 1956 a 1958, é importante observar que o números de passageiros aumenta entre os meses de julho e agosto, com uma queda até novembro e com aumento no mês de dezembro.

Julho e agosto sendo um período de férias escolares e dezembro as comemorações de Final de ano como Natal e Reveillón.
