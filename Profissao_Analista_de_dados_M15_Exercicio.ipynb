## 1\. Preços `outliers` de diamante

```
import seaborn as sns

data = sns.load_dataset("diamonds")
data.head()
```

```
# gráfico do exercício 1
import pandas as pd
import matplotlib.pyplot as plt

data["price"].describe()

plt.figure(figsize=(10, 6))
sns.boxplot(data=data, x="price", color="lightgreen")
plt.xlabel("Preço")
plt.title("Distribuição dos preços de diamantes")
plt.show()
```

**Insight do gráfico 1**:

Existem outliers significativos à direita com valores discrepantes e a mediana está concentrada na parte inferior da caixa predominando entre os valores aproximados a 2500.

## 2\. Correlação entre o do preço do diamante com seu peso (`carat`) agrupados por sua transparêcia (`clarity`)

```
import seaborn as sns

data = sns.load_dataset("diamonds")
data.head()
```

```
# gráfico do exercício 2

import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))
sns.lmplot(data=data, x="carat", y="price", hue="clarity", palette="Set2")
plt.xlabel("Peso")
plt.ylabel("Preço")
plt.title("Relação entre o preço e o peso dos diamantes")
plt.show()
```

**Insight do gráfico 2**:
Existe uma forte ligação entre o peso e o preço, os valores também variam de acordo com a clareza maior como indicado no clarity VS2 representado pela cor verde neon.

## 3\. Distribuição contínua aproximada do peso (`weight`) de carros

```
import seaborn as sns

data = sns.load_dataset("mpg")
data.head()
```

```
# gráfico do exercício 3

import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))
sns.kdeplot(data=data, x="weight", fill=True, color="purple")
plt.xlabel("Peso")
plt.ylabel("Densidade")
plt.title("Distribuição aproximada do peso dos carros")
plt.show()
```

**Insight do gráfico 3**:

A maior concentração de carros tem peso em torno de 1500 a 3500
Onde predomina a maior densidade entre 2000 a 2500.
